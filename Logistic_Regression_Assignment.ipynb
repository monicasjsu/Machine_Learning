```
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error
from sklearn.utils import resample
from google.colab import drive
import collections
```

```
#Loading Data
drive.mount('/content/drive')
data = pd.read_csv("/content/drive/My Drive/MLSpring2020/shecodes_employee_attrition/HR-Employee-Attrition.csv")
```

```
data.head()
```

```
#Data Preparation
#Checking datatype and other details of dataset
data.info()
```

```
#Checking if there are any missing values in dataset
data.isnull().sum()
```

```
#Converting categorical values to numerical
data.Attrition = pd.Categorical(data.Attrition)
data.Attrition = data.Attrition.cat.codes
data.BusinessTravel = pd.Categorical(data.BusinessTravel)
data.BusinessTravel = data.BusinessTravel.cat.codes
data.Department = pd.Categorical(data.Department)
data.Department = data.Department.cat.codes
data.EducationField = pd.Categorical(data.EducationField)
data.EducationField = data.EducationField.cat.codes
data.Gender = pd.Categorical(data.Gender)
data.Gender = data.Gender.cat.codes
data.JobRole = pd.Categorical(data.JobRole)
data.JobRole = data.JobRole.cat.codes
data.MaritalStatus = pd.Categorical(data.MaritalStatus)
data.MaritalStatus = data.MaritalStatus.cat.codes
data.Over18 = pd.Categorical(data.Over18)
data.Over18 = data.Over18.cat.codes
data.OverTime = pd.Categorical(data.OverTime)
data.OverTime = data.OverTime.cat.codes
data.head()
```

```
#Rechecking the data to make sure all features have numerical values
data.info()
```

```
# Dropping Infinites and NaN values if there are any
data.replace([np.inf, -np.inf], np.nan)
data.replace([np.inf, -np.inf],np.nan).dropna()
```

```
data['Attrition'].value_counts()
```

```
#Downsample
data_majority = data[data.Attrition==0]
data_minority = data[data.Attrition==1]
data_majority_downsampled = resample(data_majority, replace= False, n_samples= len(data_minority), random_state=123) 
data_downsampled = pd.concat([data_majority_downsampled, data_minority])
```

As we see, we have imbalanced classes. We will use sampling method to balance the classes.

```
# Check value_counts and remove any columns with singular values
single_cols = []
for col in data.columns:
    if len(data[col].unique())==1:
        print(col)
        single_cols.append(col)
# Now lets drop the columns
data.drop(single_cols, axis=1, inplace=True)
```

```
# Separate input features (X) and target variable (y)
Y = data.Attrition
X = data.drop(['Attrition'], axis=1)
```

```
# Spliting dataset into training and test data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```
#Creating instance of Logistic Regression model and fiting data in it.
LogisticR = LogisticRegression(max_iter=10000)
LogisticR.fit(x_train,y_train)
```

```
#Prediction
y_pred=LogisticR.predict(x_test)
```

```
from sklearn import metrics
ConfMatrix = metrics.confusion_matrix(y_test, y_pred)
ConfMatrix
```

Logstic model made 242 and 17 accurate predictions whereas 7 and 28 are inaccurate predictions

```
#Evaluation of the model
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

The Accuracy of our model is 88%. As per Precision, employee attrition rate would be 70%. If there are some employees who have quit their job in test data, our model can identify it 37% of the time.

*We* have implemented Logistic regression to classify values in 'Attrition' column from our dataset. We chose Attrition feature as it contains catagorical values 'Yes' and 'No' which depicts the attrition status of an employee. 
As the objective of our project is to predict employee attrition from given data, choosing Attrition column as feature will help in further analysis.


Logistic Regression computes the probability of an employee quitting his job. This model helps in solving our business problem; i.e. predicting whether an employee attrites or not.
